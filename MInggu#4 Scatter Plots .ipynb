```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
surance_filepath = "/content/drive/MyDrive/zz/insurance.csv"
surance_data = pd.read_csv(insurance_filepath)
```

```python
surance_data.head()
```

```python
sns.scatterplot(x=surance_data['bmi'], y=surance_data['charges'])
```

```python
sns.regplot(x=surance_data['bmi'], y=surance_data['charges'])
```

```python
sns.scatterplot(x=surance_data['bmi'], y=surance_data['charges'], hue=surance_data['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=surance_data)
```

```python
sns.swarmplot(x=surance_data['smoker'],
              y=surance_data['charges'])
```
